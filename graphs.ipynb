```python
import numpy as np
import matplotlib.pyplot as plt

# Example values (replace with yours)
models = ['Logistic Regression', 'XGBoost', 'KNN', 'Neural Net']

precision = [.13, .12, .12, .13]
recall    = [.91, .90, .82, .90]
accuracy  = [.80, .79, .76, .80]

x = np.arange(len(models))  # positions for models
width = 0.25  # width of each bar

plt.figure(figsize=(8,6))

plt.bar(x - width, precision, width, label='Precision')
plt.bar(x, recall, width, label='Recall')
plt.bar(x + width, accuracy, width, label='Accuracy')

plt.xticks(x, models)
plt.xlabel("Model")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.title("Model Performance Comparison", fontsize=16)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # <-- moves legend outside

plt.tight_layout()
plt.show()
```
